```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

import numpy as np 

import matplotlib as mpl

%matplotlib inline
import matplotlib.pyplot as plt 

QUERY = """ SELECT * FROM master """

df1 = pd.read_sql_query(QUERY,conn)

df1['ConvertedComp'].isnull().sum()

ConvertedComp = df1['ConvertedComp'].sort_values().median()

df1['ConvertedComp'].replace('?',np.nan,inplace=True)

df1['ConvertedComp'].replace(np.nan,ConvertedComp,inplace=True)

df1['ConvertedComp'].plot(kind='hist',figsize=(8,5))
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df1['Age'].isnull().sum()
age = df1['Age'].sort_values().median()

df1['Age'].replace('?',np.nan,inplace=True)

df1['Age'].replace(np.nan,age,inplace=True)

df1['Age'].plot(kind='box',figsize=(15,10))
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

df1[['Age','WorkWeekHrs']]

workweekhrs = df1['WorkWeekHrs'].sort_values().median()

df1['WorkWeekHrs'].replace('?',np.nan,inplace = True )

workweekhrs = df1['WorkWeekHrs'].sort_values().median()

df1['WorkWeekHrs'].replace(np.nan,workweekhrs,inplace = True )

df1['WorkWeekHrs'].isnull().sum()

df1[['Age','WorkWeekHrs']]

df1.plot(kind='scatter',x='Age',y='WorkWeekHrs',figsize=(10,5))

plt.title ( "Scatter plot between Age and WorkWeekHrs") 
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here


df1['CodeRevHrs'].sort_values()

df1['CodeRevHrs'].replace('?',np.nan,inplace=True)

c = df1['CodeRevHrs'].sort_values()

Coderevhrs = c.median()
df1['CodeRevHrs'].replace(np.nan,Coderevhrs,inplace=True)

df1[['WorkWeekHrs','CodeRevHrs','Age']]

norm_ww =((df1['Age'] - df1['Age'].min())/ (df1['Age'].max() - df1['Age'].min()))



ax0 = df1.plot(kind='scatter',x='WorkWeekHrs',y='CodeRevHrs',s=norm_ww*2000 + 10 , color='red',figsize=(20,10))

plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs')
plt.legend(['Age'],loc='upper left',fontsize='x-large')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

# pie chart of databases worked with
q2 = """SELECT * FROM DatabaseDesireNextYear"""

df2 = pd.read_sql_query(q2,conn)

data = df2.groupby('DatabaseDesireNextYear').sum()
data
df2_top5 = data.sort_values(by = 'Respondent',ascending=False)

df2_top5 = df2_top5.head()

df2_top5['Respondent'].plot(kind='pie',figsize=(15,8),autopct='%1.1f%%',startangle=90,shadow=True)


#Top 5 languageDesireNextYear

q3 = """SELECT * FROM LanguageDesireNextYear"""

df3 = pd.read_sql_query(q3,conn)
data1 = df3.groupby('LanguageDesireNextYear').count()

data1.sort_values(by='Respondent',ascending=False)


#Top 5 LanguageWorkedWith

q4 = """SELECT * FROM LanguageWorkedWith"""

df4 = pd.read_sql_query(q4,conn)

df4['LanguageWorkedWith'].value_counts()

#Top 5 DatabaseWorkedWith
q5 = """SELECT * FROM DatabaseWorkedWith"""

df5 = pd.read_sql_query(q5,conn)

df5['DatabaseWorkedWith'].value_counts()

df5.groupby(by=['Respondent']).sum()

q6 = """SELECT * FROM DatabaseWorkedWith where DatabaseWorkedWith = 'MySQL' """

df6 = pd.read_sql_query(q6,conn)

df4['LanguageWorkedWith'].value_counts()
```
